```python
# 문서로딩 청킹
# RAG: 외부지식을 활용해서 LLM 품질 높임
# 원본문서(pdf txt doc) --> 문서로딩 (loader) --> 청킹(Splitter)  --> 임베딩/저장(VectorDB)
```

```python
from langchain_core.documents import Document
doc = Document(
    page_content = '우리회사는 근속 10년이상이면 포상제도가 있습니다.',
    metadata = {
        "source": '01001.pdf',
        'page' : 1, 
        'author': '작성자',
        'created_at': '2025-11-27'
    }
)

# 문서로더 - txt
from langchain_community.document_loaders import TextLoader, PyPDFLoader, WebBaseLoader,DirectoryLoader
loader = TextLoader('document_test1.txt', encoding='utf-8')
documents = loader.load()
print (documents)
```

```python
# 문서로더 - web
loader = WebBaseLoader("https://www.mk.co.kr/news/business/11477220")
documents_web = loader.load()
print (documents_web)
```

```python
# 문서로더 - 디렉토리 로더
loader = DirectoryLoader (
    './',
    glob = "**/*.txt",
    loader_cls = TextLoader,  #loader_cls = TextLoader ==> 이거 안되면 경로 설정 loader_cls = lambda path: TextLoader(file_path=path, encoding = 'utf-8'), 로 작성하기
    show_progress=True
)
documents = loader.load()
print(documents)
```

```python
from langchain_core.documents import Document
from langchain_community.document_loaders import TextLoader, PyPDFLoader, WebBaseLoader,DirectoryLoader

# 문서로더 - 디렉토리 로더
loader = DirectoryLoader (
    './',
    glob = "**/*.pdf",
    loader_cls = PyPDFLoader,  #loader_cls = TextLoader ==> 이거 안되면 경로 설정 loader_cls = lambda path: TextLoader(file_path=path, encoding = 'utf-8'), 로 작성하기
    show_progress=True
)
documents = loader.load()
print(documents)
```
