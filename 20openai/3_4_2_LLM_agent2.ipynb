```python
from langchain.tools import tool
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain.agents import create_structured_chat_agent
```

```python
# 데코레이터

# api 설정
import os
from dotenv import load_dotenv
load_dotenv()


# 함수 설명(Docstring)과 파라미터 타입(Type Hint) ---> 명확하게 적어야 AI가 사용
@tool
def multiply(a:int, b:int) -> int :  # 타입힌트, 2개변수 int 결과물도 int ?
    '''두 정수를 곱합니다.'''
    return a*b
@tool
def get_word_length(word:str) -> int :
    '''단어의 길이를 반환합니다.'''
    return len(word)


# 도구 리스트를 생성
tools = [multiply, get_word_length]
llm = ChatOpenAI(model='gpt-4o-mini', temperature=0, api_key = os.environ["OPENAI_API_KEY"])


# 프롬프트 정의 (템플릿)
input = "단어 'langchain'의 글자수에 5를 곱하면 얼마야?"
prompt = ChatPromptTemplate.from_messages([
    ('system', "당신은 유능한 ai 어시스턴트 입니다. 주어진 도구를 사용해서 질문에 대답해 주세요"),
    ('human', input)
])
```
