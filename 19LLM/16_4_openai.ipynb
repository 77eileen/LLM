```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from openai import OpenAI
# client = OpenAI(api_key=os.environ['OPENAI_KEY'])
client = OpenAI() # 원래를 위문장으로 작성했어야 했는데, 이렇게만 해도 됨
response = client.responses.create(
    model="gpt-5-nano",
    input="서울의 날씨 알려줘"
)

print(response.output_text)
```
